```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from collections import Counter
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import gensim.downloader as api

wv = api.load('word2vec-google-news-300')
```

```python
wv
```

### 1. tokenization (做斷句)

```python
text_data = walmart_df["componenttext"].tolist()

sentences = [sent for text in text_data for sent in sent_tokenize(text)]
sentences
```

### 2. preprocess(clean data)

```python
stop_words = set(stopwords.words('english'))
def preprocess(sentence):
    words = word_tokenize(sentence.lower())
    return ' '.join([word for word in words if word.isalnum() and word not in stop_words])

preprocessed_sentences = [preprocess(sentence) for sentence in sentences]
print(f"sentence length: {len(preprocessed_sentences)}")
preprocessed_sentences
```

### 3. 建 key words 字典

```python
key_words = ["supplier", "risk", "inventory", "customer"]
similar_words = []

for key_word in key_words:
    try:
        similar = wv.most_similar(key_word, topn=20)
        similar_words.extend([word for word, _ in similar])
    except KeyError:
        print(f"Warning: '{key_word}' not found in the vocabulary")
        
dictionary = set(similar_words + key_words)
dictionary
```

### 4. 找尋每個句子內有幾個 keywords

```python
sentence_scores = []
for sentence in preprocessed_sentences:
    words_in_sentence = word_tokenize(sentence)
    count = sum(1 for word in words_in_sentence if word in dictionary)
    sentence_scores.append((count, sentences[preprocessed_sentences.index(sentence)]))
sentence_scores
```

```python
top_sentences = sorted(set(sentence_scores), key=lambda x: x[0], reverse=True)[:5]
```

```python
print("Top sentences:")
for score, sentence in top_sentences:
    print(f"Score: {score}, Sentence: {sentence}")
```

```python
# Analyze frequently occurring words
all_words = [word for sentence in preprocessed_sentences for word in word_tokenize(sentence) if word in dictionary]
word_freq = Counter(all_words)
print("\nMost common words:")
print(word_freq.most_common(10))
```

```python
# Print similar words for each key word
for key_word in key_words:
    print(f"\nSimilar words to '{key_word}':")
    try:
        similar = wv.most_similar(key_word, topn=20)
        print(", ".join([word for word, _ in similar]))
    except KeyError:
        print(f"'{key_word}' not found in the vocabulary")
```

### 5. 使用 sbert 找出和種子句 embedding 相似的句子

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
sentence_embeddings = model.encode(preprocessed_sentences)
print(f"sentence embeddings length: {len(sentence_embeddings)}")
```

```python
seed_sentence = top_sentences[0][1]
seed_embedding = model.encode([seed_sentence])[0]
```

```python
similarities = cosine_similarity([seed_embedding], sentence_embeddings)[0]
n = 5  
top_n_indices = similarities.argsort()[-n-1:-1][::-1]  # 找到相似度最高的前 n 個句子，並排除種子句自己
top_n_sentences = [sentences[i] for i in top_n_indices]

for idx, sentence in enumerate(top_n_sentences):
    print(f"Top {idx+1} similar sentence: {sentence}")
```
